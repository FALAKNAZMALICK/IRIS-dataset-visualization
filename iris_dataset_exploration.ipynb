# Task 1: Iris Data Exploration
In this task, I will perform Exploratory Data Analysis (EDA) on the Iris dataset. The goal is to load the data, summarize its structure, and create visualizations to understand the relationships between flower features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset load
df = sns.load_dataset('iris')
print("Data Load !")

```

### Dataset Overview * Shape: The dataset contains 150 rows and 5 columns.
* Features: The columns include sepal length, sepal width, petal length, petal width, and the species name

```python
print(df.shape)      # Rows aur columns
print(df.columns)    # Columns name
print(df.head())     # Start 5 rows
```

### Analysis: Feature Relationships

The scatter plot shows a clear separation between the Setosa species and the others.

There is a strong positive correlation between petal length and petal width; as one increases, the other tends to increase as well.

```python
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title("Scatter Plot: Sepal Length vs Width")
plt.show()
```

### Analysis: Data Distribution

The histogram shows that petal lengths are not normally distributed.

There is a large gap between small petals (around 1-2cm) and larger petals (3-7cm), indicating different groups within the data.

```python
sns.histplot(df['petal_length'], kde=True)
plt.title("Histogram: Petal Length Distribution")
plt.show()
```

### Analysis: Outliers and Spread * The box plot helps identify the median and spread of each species.

Virginica has the largest spread in petal width, while Setosa has the smallest.
* Any individual points outside the "whiskers" of the plot represent potential outliers.

```python
sns.boxplot(x='species', y='petal_width', data=df)
plt.title("Box Plot: Petal Width by Species")
plt.show()
```

### Conclusion Through this exploration, I have successfully loaded and visualized the Iris dataset using pandas, matplotlib, and seaborn. Key insights include the distinct clustering of the Setosa species and the high correlation between petal measurements.
